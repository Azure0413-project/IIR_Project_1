```python
import re

def extract_number(response):
    # 提取 </think> 之後的內容
    content_after_think = response.split('</think>')[-1].strip()
    
    # 嘗試匹配被 $ 包住的數字
    match_dollar = re.search(r'\$(.*?)\$', content_after_think)
    if match_dollar:
        return match_dollar.group(1).strip()  # 返回去掉 $ 的數字
    
    # 若無 $，則匹配被 </think> 和 : 包住的數字
    match_colon = re.search(r'^([^:]+):', content_after_think)
    if match_colon:
        return match_colon.group(1).strip()
    
    return None  # 若找不到符合條件的數字，返回 None
```

```python
# 字串比對
base_correct_cnt = 0
finetuned_correct_cnt = 0
total_number = len(compared_result)
print(f"Total number of testing math queries: {total_number}")
print(f"Total number of ground truth math queries: {len(gt)}")

for item in zip(compared_result, gt[:total_number]):
    base_compared = item[0]['base_answer']
    finetuned_compared = item[0]['fine_tuned_answer']
    finetuned_answer = extract_number(finetuned_compared)
    gt_ans = str(item[1]['answer'])

    print(f"Finetuned answer: {finetuned_answer} / GT answer: {gt_ans}")

    if str(base_compared) == gt_ans:
        base_correct_cnt += 1
    #print(str(finetuned_compared))
    #print("====")
    if str(finetuned_answer) == gt_ans:
        finetuned_correct_cnt += 1

print(base_correct_cnt, finetuned_correct_cnt)
print("Base accuracy: {:.2f}".format(base_correct_cnt / total_number))
print("Finetuned accuracy: {:.2f}".format(finetuned_correct_cnt / total_number))
```
